# Data distribution

## Skewness
Skewness is the measure of how a data distribution deviates from being symmetrical around its mean. A data distribution is considered skewness if most of the data is concentrated on one side of the mean, while the remaining data is spread out on the other side. To interpret the skewness index, you should take the absolute value (remove the positive or negative sign) and interpret it as follows:

    Skewness index < 0.15 = symmetrical distribution
    Skewness index between 0.15 and 1.0 = moderately asymmetric distribution
    Skewness index > 1 = strongly asymmetric distribution

## Kurtosis
Kurtosis is the measure of how flat or peaked a data distribution is compared to a standard distribution, known as the normal curve. The normal distribution has a kurtosis of 0.263.

Source: [UFMG](https://www.bu.ufmg.br/imagem/00002c/00002cd9.pdf) translate with ChatGPT

```python
cols=['Hours per day','Age']
for col in cols:
    print('----------------------------------------------')
    fig = plt.figure(figsize =(20, 5))
    print(f'Kurtosis of {col}: {df[col].kurtosis():.2f}\nSkewness of{col}: {df[col].skew():.2f}')
    sns.histplot(df[col])
    plt.title(f'Distribuition of {col}')
    
```

# Hypothesis testing

### Normality Test

```python
## Normality
import scipy.stats as stats
pvalue=0.05

for col in cols:
    print('----------------------------------------------')   
    print(f'According to D’Agostino and Pearson’s  normality test, the distribution of the column {col} {"is normal" if stats.normaltest(df[col]).pvalue > pvalue else "is not normal"}') 
```

### Mann-Whitney

```python
for categorical in categories:
    for colname in cols:
        group1=df[(df[categorical]==df[categorical].unique()[0])][colname]
        group2=df[(df[categorical]==df[categorical].unique()[1])][colname]
        res = stats.mstats.mannwhitneyu(group1, group2)
        print(f'The Mann-Whitney Test using the medians of {colname} concerning {categorical} has a p-value of {res.pvalue:.3f}. In other words, the medians {"are different" if res.pvalue < pvalue else "are the same"}.')
        print('-------------------------------------------------------------------------------------------------------------------------------------------------')
```

# Kruskal-Wallis

```python
genres=['Frequency [Classical]','Frequency [Country]', 'Frequency [EDM]', 'Frequency [Folk]',
       'Frequency [Gospel]', 'Frequency [Hip hop]', 'Frequency [Jazz]','Frequency [K pop]', 'Frequency [Latin]', 'Frequency [Lofi]',
       'Frequency [Metal]', 'Frequency [Pop]', 'Frequency [R&B]','Frequency [Rap]', 'Frequency [Rock]', 'Frequency [Video game music]']

cols=['Hours per day','Age']

for genre in genres:
    scale=len(df[genre].unique())
    list_genre=[]
    for colname in cols:
        for i in range(scale):
            subset=df[(df[genre]==df[genre].unique()[i])][colname]
            list_genre.append(subset.tolist())
            
        res = stats.mstats.kruskal(*list_genre)
        print(f'The Kruskal-Wallis Test when comparing whether there is a difference between the medians of {genre} with {colname} obtained a p-value of {res.pvalue:.3f}. This indicates that {"there is a difference" if res.pvalue < 0.05 else "there is no difference"} between the categories.')
        print('-------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------')
        
```
